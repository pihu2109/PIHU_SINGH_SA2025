### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
#your code
filtered_cars = cars[(cars['horsepower'] > 100) & (cars['weight'] < 3000)]
most_common_origin = filtered_cars['origin'].mode()[0]
print(f"The most common origin is: {most_common_origin}")
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
#your code here
mean_accel = cars[cars['origin'] == 'japan']['acceleration'].mean()
print(f"Mean acceleration of cars from Japan: {mean_accel:.2f}")
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
median_mpg=cars['mpg'].mean()
above_median=cars[cars['mpg']>median_mpg]
max_hp_per_weight=above_median.index.max()
best_above = above_median.loc[[max_hp_per_weight]] 
print(best_above['name'])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
import seaborn as sns
# Create the multi-line plot
mean_mpg = cars.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.lineplot(data=mean_mpg, x='model_year', y='mpg', hue='origin', marker='o')

# Add titles and labels
plt.title('Average MPG Over the Years by Origin')
plt.xlabel('model_year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
import seaborn as sns
import matplotlib.pyplot as plt
cars = cars.reset_index()
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=cars,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    sizes=(40, 200),  # Min and max point sizes
    hue_order=['japan', 'europe', 'usa'],
    palette='Set1',
    alpha=0.7
)
plt.title('Horsepower vs. Weight of Cars by Origin (Sized by MPG)')
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.legend(title='Origin', bbox_to_anchor=(1.05, 1))  # Move legend outside
plt.tight_layout()
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
import pandas as pd
name_year_counts = cars.groupby('name')['model_year'].nunique()
multi_year_names = name_year_counts[name_year_counts > 1].index
multi_year_cars = cars[cars['name'].isin(multi_year_names)]
consistency_stats = multi_year_cars.groupby('name').agg(
    count=('mpg', 'count'),
    mpg_std=('mpg', 'std'),
    mpg_avg=('mpg', 'mean')
).reset_index()
consistent_cars = consistency_stats[consistency_stats['mpg_std'] < 1.0]
consistent_cars = consistent_cars.sort_values(by=['count', 'mpg_avg'], ascending=[False, False])
print(consistent_cars[['name', 'count', 'mpg_avg']])
```
